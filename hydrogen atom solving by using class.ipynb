#### radius of nth orbit

$r_n=\frac{n^2h^2\epsilon_0}{\pi me^2}$ =$(0.53)n^2 A$

##### velocity of electron in nth orbit

$v_n=\frac{e^2}{2nh\epsilon_0}$ =$\frac{2.2*10^6}{n} m/s $

##### Total energy of electron in nth orbit

$E_n = \frac{-me^4}{8\epsilon_0^2 n^2h^2}$ = $\frac{-13.6 eV}{n^2}$

```python
def Hydrogen_radius(n):
    r = n**2*k
    return r
k = 0.53
Hydrogen_radius(1),Hydrogen_radius(2),Hydrogen_radius(3),Hydrogen_radius(4)   
```

```python
def electron_velocity(n):
    v = k1/n
    return v
k1 = 2.2*10**6
electron_velocity(1),electron_velocity(2),electron_velocity(3),electron_velocity(4)
    
```

```python
def Total_energy(n):
    E = k2/n**2
    return E
k2 = -13.6
Total_energy(1),Total_energy(2),Total_energy(3),Total_energy(4)
```

###### To create a class for hydrogen atom

```python
class Hydrogen_atom():
    def __init__(self,value_k,value_k1,value_k2,info):
        self.k=value_k
        self.k1=value_k1
        self.k2=value_k2
        self.info=info
    def Hydrogen_radius(self,n):
        rn=self.k*n**2
        return rn
    def electron_velocity(self,n):
        vn=self.k1/n
        return vn
    def Total_energy(self,n):
        En=self.k2/n**2
        return En
H=Hydrogen_atom(value_k=0.53,value_k1=2.2*10**6,value_k2=-13.6,info='Wow')
H.Hydrogen_radius(1),H.electron_velocity(1),H.Total_energy(1)
```

```python
N=[]
R=[]
V=[]
E=[]
for n in range(1,11):
    rr=H.Hydrogen_radius(n)
    vv=H.electron_velocity(n)
    ee=H.Total_energy(n)
    N.append(n)
    R.append(rr)
    V.append(vv)
    E.append(ee)
   # print(N,R,V,E)
    
```

```python
data={}
data.update({"principalquantumnumber":N,"radius":R,"velocity":V,"energy":E})
print(data)
```

```python
import pandas as pd
DF=pd.DataFrame(data)
print(DF)
```

```python
DF.to_csv("Hydrogen_atom.csv")
DF.head()
```
